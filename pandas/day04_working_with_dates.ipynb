```python
#Import pandas 
import pandas as pd
```

```python
#Load dataset
df = pd.read_csv(r"C:\Users\marya\OneDrive\Desktop\fintech\finpay-data-analytics-roadmap\pandas\customers_clean.csv")
```

```python
#Displays the dataframe 
df
```

```python
df.info()
```

```python
#Convert signup date to datetime data type
df["signup_date"] = pd.to_datetime(df["signup_date"])
```

```python
df.info()
```

```python
#Extract the signup year
df["signup_year"] = df["signup_date"].dt.year
```

```python
#Extract the signup month
df["signup_month"] = df["signup_date"].dt.month_name()
```

```python
#Extract the day
df["signup_day"] = df["signup_date"].dt.day
```

```python
#Extract weekday
df["signup_weekday"] = df["signup_date"].dt.day_name()
```

```python
#Extract the quarter
df["signup_quarter"] = df["signup_date"].dt.quarter
```

```python
#Calculate Days since signup
#Today's date - Signup Date = Number of Days
df["days_since_signup"] = (
    pd.Timestamp.today()- df["signup_date"]
).dt.days

df
```

```python
#Which month had the most signups?
df["signup_month"].value_counts()
```

```python
#Which weekday sees the highest signup volume?
df["signup_weekday"].value_counts()
```

```python
#Are signups increasing over time?
df.groupby("signup_year").size()
```
