```python
Import pandas
import pandas as pd
```

```python
#Load the dataset
df = pd.read_csv(r"C:\Users\marya\OneDrive\Desktop\fintech\finpay-data-analytics-roadmap\pandas\customers_features.csv")
```

```python
#Displays the dataframe
df
```

```python
#Which city has the largest customer base?
df["city"].value_counts()
```

```python
#Which city has the highest average annual income?
df.groupby("city")["annual_income"].mean().sort_values(ascending=False)
```

```python
#Which employment status has the highest average income?
df.groupby("employment_status")["annual_income"].mean().sort_values(ascending=False)
```

```python
#Which age group has the highest average credit score?
df.groupby("age_group")["credit_score"].mean().sort_values(ascending=False)
```

```python
#How many customers belong to each risk segment?
df["risk_segment"].value_counts()
```

```python
#Which city has the most high-risk customers?
df[df["risk_segment"]=="High Risk"]["city"].value_counts()
```

```python
#How many high-value customers do we have?
df["high_value_customer"].value_counts()
```

```python
#Which city has the most high-value customers?
df[df["high_value_customer"]=="Yes"]["city"].value_counts()
```

```python
#Which income band contains the most customers?
df["income_band"].value_counts()
```

```python
#What is the average annual income by age group?
df.groupby("age_group")["annual_income"].mean().sort_values(ascending=False)
```

```python
#Which signup quarter had the most customers?
df["signup_quarter"].value_counts()
```

```python
#Which weekday had the highest number of customer signups?
df["signup_weekday"].value_counts()
```
