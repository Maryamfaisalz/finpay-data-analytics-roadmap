```python
#Import Library
import pandas as pd
```

```python
#Load the customers_dirty.csv dataset into a DataFrame.
df = pd.read_csv(r"C:\Users\marya\OneDrive\Desktop\fintech\finpay-data-analytics-roadmap\datasets\customers_dirty.csv")

#Display the dataframe
df
```

Basic Outlier Detection

```python
#Calculate Q1 and Q3 and IQR
Q1 = df["annual_income"].quantile(0.25)
Q3 = df["annual_income"].quantile(0.75)

IQR = Q3 - Q1

#Lower bound
lower = Q1 - 1.5*IQR 

#Upper bound
upper = Q3 + 1.5*IQR
```

```python
#Outliers
df[(df["annual_income"]<lower) | (df["annual_income"]>upper)]
```

```python
#Check if there are any null values left
df.isnull().sum()
```

```python
#Convert to numeric datatype
df["credit_score"] = pd.to_numeric(df["credit_score"],errors="coerce")
```

```python
#Remove rows where the signup date is missing
df = df.dropna(subset=["signup_date"])
```

```python
#Check the signup date still has missing values
df["signup_date"].isnull().sum()
```
