```python
#Import pandas and numpy 
import pandas as pd
```

```python
#Load dataset 
df = pd.read_csv(r"C:\Users\marya\OneDrive\Desktop\fintech\finpay-data-analytics-roadmap\pandas\customer_dates.csv")
```

```python
#Displays the dataframe
df
```

```python
#creates a new column showing each customer's estimated monthly income
df["monthly_income"] = df["annual_income"]/12
```

```python
#Clarifies customers as high value based on whether their annual income exceeds 100,000
df["high_value_customer"] = np.where(df["annual_income"]>100000,"Yes","No")
```

```python
#Converts annual income into thousands for easier reporting
df["annual_income_k"] = df["annual_income"].apply(lambda x:round(x/1000,1))
```

```python
#Maps each risk level to a numerical score for easier analysis
df["risk_score"] = df["risk_level"].map({"Low":1,"Medium":2,"High":3})
```

```python
#Groups customers in age categories
def age_group(age):
    if age < 25:
        return "Young Adult"
    elif age < 40:
        return "Adult"
    elif age < 60:
        return "Middle Aged"
    else:
        return "Senior"
```

```python
df["age_group"] = df["age"].apply(age_group)
```

```python
#Categorizes customers into income bands
def income_band(income):
    if income < 50000:
        return "Low"
    elif income < 100000:
        return "Medium"
    else:
        return "High"
```

```python
df["income_band"] = df["annual_income"].apply(income_band)
```

```python
#Categorizes customers based on their credit score 
def credit_category(score):
    if score >= 750:
        return "Excellent"
    
    elif score >=650:
        return "Good"

    elif score >= 550:
        return "Fair"

    else:
        return "Poor"
```

```python
df["credit_score_category"] = df["credit_score"].apply(credit_category)
```

```python
#Renames the tenure column to improve readability
df.rename(columns={
    "days_since_signup":"customer_tenure_days"
},inplace=True)
```

```python
#Identifies customers with both high income and excellent credit
df["high_value_customer"] = np.where(
    (df["annual_income"]>100000) & (df["credit_score"]>750),"Yes","No"
)
```

```python
#Converts risk level into business friendly risk segments
df["risk_segment"] = df["risk_level"].map({"Low":"Safe","Medium":"Monitor","High":"High Risk"})
```

```python
#Which age group has the most of the customers?
df["age_group"].value_counts()
```

```python
#How many high value customers do we have?
df["high_value_customer"].value_counts()
```

```python
#Which income band has the most customers?
df["income_band"].value_counts()
```

```python
#Which credit score category is most common?
df["credit_score_category"].value_counts()
```

```python
#Average income by age group
df.groupby("age_group")["annual_income"].mean().sort_values(ascending=False)
```

```python
#Average credit score by income band 
df.groupby("income_band")["credit_score"].mean().sort_values(ascending=False)
```
