```python
#Import pandas 
import pandas as pd 
```

```python
#Load the merged of transaction and merchant dataset
df = pd.read_csv(r"C:\Users\marya\OneDrive\Desktop\fintech\finpay-data-analytics-roadmap\datasets\transactions_merchant_master.csv")
df
```

Executive KPI's

```python
#Total Revenue
total_revenue = df["amount"].sum().round(2)
print("Total Revenue:", total_revenue)
```

```python
#Total transactions
total_transactions = df["transaction_id"].count()
print("Total Transactions:",total_transactions)
```

```python
#Average transaction value
average_transaction = df["amount"].mean().round(2)
print("Average Transaction:",average_transaction)
```

```python
#Highest Transaction 
highest_transaction = df["amount"].max()
print("Highest Transaction:",highest_transaction)
```

```python
#Lowest Transaction 
lowest_transaction = df["amount"].min()
print("Lowest Transaction:",lowest_transaction)
```

```python
#Total cashback 
total_cashback = df["cashback"].sum()
print("Total Cashback:", total_cashback)
```

```python
#Average Cashback
average_cashback = df["cashback"].mean().round(2)
print("Average Cashback:",average_cashback)
```

```python
#Number of merchants 
number_of_merchants = df["merchant_id"].nunique()
print("Number of Merchants:",number_of_merchants)
```

```python
#Number of Merchant categories 
merchant_categories = df["category"].nunique()
print("Number of Merchant Categories:",merchant_categories)
```

```python
#Number of Payment Method
payment_methods = df["payment_method"].nunique()
print("Number of Payment Methods:",payment_methods)
```

```python
#Transaction success rate
successful_transactions = df["transaction_status"]=="Success"
total_transactions = len(df["transaction_status"])
success_rate = (successful_transactions.sum()/total_transactions)*100
print("Success Rate:",success_rate,"%")
```

```python
#Failed transaction rate 
failed_transaction = df["transaction_status"]=="Failed"
total_transactions = len(df["transaction_status"])
failed_rate = (failed_transaction.sum()/total_transactions)*100
print("Failed Rate:",failed_rate,"%")
```

```python
#Pivot tables 1 - Revenue by Merchant Category 
pd.pivot_table(df,values="amount",index="category",aggfunc="sum")
```

```python
#Revenue by Payment Method 
pd.pivot_table(df,values="amount",index="payment_method",aggfunc="sum")
```

```python
#Transaction Count by Merchant Category
pd.pivot_table(df,values="transaction_id",index="category",aggfunc="count")
```

```python
#Average transaction Value
pd.pivot_table(df,values="amount",index="category",aggfunc="mean")
```

```python
#Cashback by Merchant Category
pd.pivot_table(df,values="cashback",index="category",aggfunc="sum")
```

```python
#Revenue by Merchant x Payment Method 
pd.pivot_table(df,values="amount",index="category",columns="payment_method",aggfunc="sum")
```

```python
#Transaction status by payment method
pd.pivot_table(df,values="transaction_id",index="payment_method",columns="transaction_status",aggfunc="count",fill_value=0)
```

```python
#Merchant Performance dashboard
pd.pivot_table(df,values="amount",index="merchant_name",aggfunc=["sum","mean","count"])
```
