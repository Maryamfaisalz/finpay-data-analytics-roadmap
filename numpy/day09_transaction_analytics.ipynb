Payment Analytics

```python
#Which payment method are available?
np.unique(payment)
```

```python
#How many transactions were made using each payment method?
methods,counts = np.unique(payment,return_counts=True)

for method,count in zip(methods,counts):
    print(method,count)
```

```python
#Which payment method is the most popular?
methods[np.argmax(counts)]
```

```python
#Which payment method is the least popular?
methods[np.argmin(count)]
```

```python
#What percentage of transactions used each payment method?
percentages = (counts/len(payment))*100
percentages

for method,pct in zip(methods,percentages):
    print(method,round(pct,2))
```
