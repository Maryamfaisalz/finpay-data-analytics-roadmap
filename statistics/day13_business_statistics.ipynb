Outlier Detection

```python
#Are there any unusually large transactions that require investigation?
Q1 = df["amount"].quantile(0.25)
Q1
```

```python
Q3 = df["amount"].quantile(0.75)
Q3
```

```python
#Interquartile Range
IQR =Q3 - Q1
IQR
```

```python
#Finding lower and upper bound
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
lower
```

```python
upper
```

```python
#Finding outliers
outliers = df[(df["amount"]<lower)|(df["amount"]>upper)]
outliers
```

There seems to be no outliers according to the IQR method. 

```python
#checking if there are any outliers
len(outliers)
```

```python
#Checking the shape of the outlier
outliers.shape
```

Business Insight:
- No statistical outliers were detected in transaction amounts using the IQR method
- Customer spending falls within the expected range of the dataset
- There are no unusually high or unusually low transactions requiring further investigation

Business Actions:
- Continue monitoring transaction amount as new data is collected 
- Apply the same outlier detection process regularly to identify unusual spending patterns or potential fradulent transactions
