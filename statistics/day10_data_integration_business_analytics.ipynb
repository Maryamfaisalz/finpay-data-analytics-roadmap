Transaction Analytics

```python
#How many transactions were successful and failed?
transactions_merchants["transaction_status"].value_counts()
```

```python
#What is the transaction success rate?
successful_transaction = transactions_merchants["transaction_status"]=="Success"
total_transactions = len(transactions_merchants["transaction_status"])
transaction_success_rate = successful_transaction.sum()/total_transactions*100
transaction_success_rate
```
